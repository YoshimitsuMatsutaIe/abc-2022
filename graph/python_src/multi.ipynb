## 複数のグラフを並べて表示
シミュレーション結果をまとめて表示するときに使う

````python
import matplotlib.pylab as plt
import numpy as np
````

データ用意

````python
x1 = np.arange(-1, 1, 0.01)
y1_1 = np.sin(x1); y1_2 = np.cos(x1)

x2 = np.arange(0, 5, 0.01)
y2_1 = np.exp(x2); y2_2 = np.exp(x2) * np.sin(x2)
````

add_subplotの引数で指定
```
add_subplot(縦の枚数, 横の枚数, 全体での位置)
```

````python
fig = plt.figure(figsize=(8, 6))
ax = fig.add_subplot(2, 2, 1)
ax.plot(x1, y1_1, label="sin")
ax.legend()

ax2 = fig.add_subplot(2, 2, 2)
ax2.plot(x1, y1_2, label="cos")
ax2.legend()

ax3 = fig.add_subplot(2, 2, 3)
ax3.plot(x2, y2_1, label="exp")
ax3.legend()

ax4 = fig.add_subplot(2, 2, 4)
ax4.plot(x2, y2_2, label="exp*sin")
ax4.legend()

plt.show()
````


一行で書くことも可能

````python
fig, axes = plt.subplots(nrows=2, ncols=2, figsize=(8, 6))
axes[0, 0].plot(x1, y1_1, label="sin")
axes[0, 1].plot(x1, y1_2, label="cos")
axes[1, 0].plot(x2, y2_1, label="exp")
axes[1, 1].plot(x2, y2_2, label="exp*sin")
for ax in axes.ravel():
    ax.legend()
plt.show()
````
