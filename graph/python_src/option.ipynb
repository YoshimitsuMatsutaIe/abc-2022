# Matplotlibのオプション一覧
よく使う線のスタイルやマーカー，カラー等のオプションをまとめました．  

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
# データ
x = np.arange(0, 2*np.pi, 0.5)
y = [np.sin(x+i) for i in np.arange(0, np.pi, np.pi/6)]
```

## linestyle

`linestlle = 名前 or 記号 or タプル`で指定する．
単純な線種は`"solid"`， `"dotted"`， `"dashed"`，`"dashdot"`の4種類．  
名前以外にも記号で指定できる．`'_'`, `':'`，`'__'`，`'_.'`と書いても良い．  

公式ページ : [https://matplotlib.org/stable/gallery/lines_bars_and_markers/linestyles.html](https://matplotlib.org/stable/gallery/lines_bars_and_markers/linestyles.html)

```python
plt.plot(x, y[0], linestyle='solid', label='solid')  # same as '_'
plt.plot(x, y[1], linestyle='dotted', label='dotted')  # same as ':'
plt.plot(x, y[2], linestyle='dashed', label='dashed')  # same as '__'
plt.plot(x, y[3], linestyle='dashdot', label='dashdot')  # same as '_.'
plt.legend()
```

## marker

多すぎるので使ったことのあるものだけ紹介．  

公式ページ : [https://matplotlib.org/stable/gallery/lines_bars_and_markers/marker_reference.html#unfilled-markers](https://matplotlib.org/stable/gallery/lines_bars_and_markers/marker_reference.html#unfilled-markers)

```python
plt.plot(x, y[0], marker='.', label='.', markersize=11)
plt.plot(x, y[1], marker='+', label='+', markersize=11)
plt.plot(x, y[2], marker='o', label='o', markersize=11)
plt.plot(x, y[3], marker='^', label='^', markersize=11)
plt.plot(x, y[4], marker='s', label='s', markersize=11)
plt.legend()
```
