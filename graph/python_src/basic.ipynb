# 基本
ここではMatplotlibの基本的なプロットを扱う．  
公式ページがわかりやすいので疑問があったらそちらを参照．  

公式ページ：[https://matplotlib.org/stable/plot_types/index.html](https://matplotlib.org/stable/plot_types/index.html)

````python
# 必要なものをインポート
import matplotlib.pylab as plt
import numpy as np  # グラフに使うデータ作成に使う
````

## plot

まずはデータ作成

````python
x = np.arange(0, 2*np.pi, 0.01)
y = np.sin(x)
````

### MATLABチックな方法
MATLABに似た文法で作る方法．  
```
plt.plot(x軸のデータ，y軸のデータ)  
```
でプロット作成できる．

````python
plt.plot(x, y)
plt.show()  # 別ウィンドウにインタラクティブなグラフを出す．jupyterでは不要．
````

### オブジェクト指向インターフェイスを使う方法
オブジェクトを使う方法です．作成者がこっちに慣れてるので以降はこの流儀でいきます．

````python
fig = plt.figure()
ax = fig.add_subplot(111)
ax.plot(x, y)
plt.show()
````

#### グラフのサイズ，ラベル，凡例，軸範囲等の設定  
* グラフのサイズ  
plt.figureの引数figsizeで指定  
* ラベル  
ax.plot()の引数labelで追加  
* 軸ラベル  
ax.set_xlabel(ラベル名)  
* 軸範囲  
ax.set_xlim(下限，上限)  
* グリッド線追加  
ax.grid()  
* 凡例表示  
ax.legend()

````python

fig = plt.figure(figsize=(9, 3))
ax = fig.add_subplot(111)
ax.plot(x, y, label="sin")  # ラベルは引数のlabelで指定
ax.set_xlabel("x")  # 軸ラベル
ax.set_ylabel("y")
ax.set_xlim(0, 2*np.pi)  # 軸範囲
ax.set_ylim(-1.1, 1.1)
ax.grid()  # グリッド線
ax.legend()  # 凡例をつける
plt.show()
````

#### グラフの軸スケールを揃える
```set_aspect('equal')```で軸スケールが揃えられる．  
ただしset_xlimやfigsizeを追加で指定する場合，グラフが正方形にならないと軸スケールは揃わない．  

````python
fig = plt.figure()
ax = fig.add_subplot(111)
ax.plot(x, y)
ax.set_aspect('equal')
plt.show()
````

#### グラフの保存
fig.savefigで保存できる 

````python
fig.savefig("basic.png")
````

#### 複数データのプロット
簡単に追加できる  

````python
y2 = np.cos(x)  # 追加したいデータ

fig = plt.figure()
ax = fig.add_subplot(111)
ax.plot(x, y, label="sin")
ax.plot(x, y2, label="cos")  # axに追加するだけ
ax.legend()
plt.show()
````

#### 色，マーカー，ラインスタイル等の設定
ax.plotの引数で指定できる  
* linestyle : 線の種類  
* color : 色  
* marker : マーカー  
  
他にもマーカーサイズやマーカーの透明度，線の太さなど設定できる項目はある．  

````python
x = np.arange(0, 2*np.pi, 0.5)
y = np.sin(x); y2 = np.cos(x)

fig = plt.figure()
ax = fig.add_subplot(111)
ax.plot(x, y, label="sin", linestyle="solid", color="black", marker="o")
ax.plot(x, y2, label="cos", linestyle="dashed", color="red", marker="*")
ax.legend()
plt.show()
````

## scatter
散布図にはax.scattrを使う  
まずはデータ作成

````python
x = np.random.rand(30)
y = np.random.rand(30); y2 = np.random.rand(30)
````

````python
fig = plt.figure()
ax = fig.add_subplot(111)
ax.scatter(x, y, label="1")
ax.scatter(x, y2, label="2")
ax.legend()
plt.show()
````

plotと同じようにマーカーの種類やサイズを変更できる

````python
fig = plt.figure()
ax = fig.add_subplot(111)
ax.scatter(x, y, label="1", marker="o", color="red", s=30, alpha=0.5)
ax.scatter(x, y2, label="2", marker="^", color="black", s=30)
ax.legend()
plt.show()
````
