# アニメーション
matplotlib.animationよりも簡単につくれる

````julia
using Plots
````

データを準備

````julia
x = 0:0.1:2π
y = sin.(x);
````

```@animate```マクロを使う．  
forループの中で作られたfigが1フレームになる  

````julia
anim = @animate for i in 1:length(x)
    fig = plot(xlim=(0, 2π))
    scatter!(fig, [x[i]], [y[i]], label="moving point")
    plot!(fig, x, y, label="line")
end
````

`gif`でアニメーションをgifに変更

````julia
gif(anim, "anim.gif", fps=10)
````
