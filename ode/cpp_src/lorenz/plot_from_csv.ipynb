# グラフ化
計算したデータをグラフで確認する．

```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

****
## ローレンツ方程式
読み込み

```python
data_lorenz = pd.read_csv('rk.csv')
data_lorenz.head()
```

プロット

```python
fig = plt.figure(figsize=(8, 8))
ax = fig.add_subplot(projection='3d')
ax.plot(data_lorenz['x'], data_lorenz['y'], data_lorenz['z'])
ax.set_xlabel("x")
ax.set_ylabel("y")
ax.set_zlabel("z")
```
